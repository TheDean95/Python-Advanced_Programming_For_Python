```python
Agency_Expenditures_NYC = pd.read_csv ('AgencyExpenditures.csv', index_col='Year')
Major_Felonies_NYC = pd.read_csv ('SevenFelonies_NYC.csv', index_col='Year')
```

```python
color1 = 'tab:red'
color2 = 'tab:blue'
color3 = 'tab:orange'
```

```python
# new dataframe

new_table = pd.concat([Agency_Expenditures_NYC[20:41],Major_Felonies_NYC], axis=1)
new_table = new_table.reset_index()
```

```python
new_table
```

```python
a = new_table.columns[1]+' vs '+new_table.columns[2]
a
```

```python
#function to draw the picture. <i> and <j> are the number of the columns of new_table respectively. 
#for example, i=1 that is Education

def draw_fig(i,j):
    fig, ax = plt.subplots(1,1, figsize=(8,6))
    ax.set_xlabel('Year',fontsize=15)
    ax.set_ylabel(new_table.columns[i],color=color1,fontsize=15)
    ax.plot(new_table.iloc[:,0], new_table.iloc[:,i], color=color1)
    
    ax2 = ax.twinx()
    ax2.set_ylabel(new_table.columns[j], color=color2,fontsize=15)
    ax2.plot(new_table.iloc[:,0],new_table.iloc[:,j], color=color2)
    ax.set_title(new_table.columns[i]+' vs '+new_table.columns[j],fontsize=20)
    
    ax.xaxis.set_major_formatter(formatter)
    ax.yaxis.set_major_formatter(formatter)
    ax2.yaxis.set_major_formatter(formatter)
    
    plt.show()
```
